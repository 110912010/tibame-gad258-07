```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#  CSV
df_red = pd.read_csv('winequality-red.csv', sep=';')

correlation_matrix = df_red.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

```

```python
#  CSV
df_white = pd.read_csv('winequality-white.csv', sep=';')

correlation_matrix = df_white.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# 
cols_to_plot = ['total sulfur dioxide', 'density', 'pH','sulphates', 'alcohol']

#  3x2 
fig, axes = plt.subplots(3, 2, figsize=(12, 10))

#  3x2 
axes = axes.flatten()

for i, col in enumerate(cols_to_plot):
    sns.regplot(x='quality', y=col, data=df_red, ax=axes[i], 
                scatter_kws={'alpha':0.3}, line_kws={'color':'orange'})
    axes[i].set_title(f'{col} vs Quality')

plt.tight_layout()
plt.show()
```
