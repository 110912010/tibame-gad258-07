```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 讀取紅酒 CSV
df_red = pd.read_csv('winequality-red.csv', sep=';')

correlation_matrix = df_red.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

```

```python
# 讀取白酒 CSV
df_white = pd.read_csv('winequality-white.csv', sep=';')

correlation_matrix = df_white.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# 挑選欄位
cols_to_plot = ['total sulfur dioxide', 'density', 'pH','sulphates', 'alcohol']

# 建立 2x3 的畫布
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

# 將 2x3 的矩陣展開成一維，方便迴圈讀取
axes = axes.flatten()

for i, col in enumerate(cols_to_plot):
    sns.regplot(x='quality', y=col, data=df_red, ax=axes[i], 
                scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
    axes[i].set_title(f'{col} vs Quality')

plt.tight_layout()
plt.show()
```

```python
# 挑選欄位
cols_to_plot = ['total sulfur dioxide', 'density', 'pH','sulphates', 'alcohol']

# 建立 2x3 的畫布
fig, axes = plt.subplots(2,3, figsize=(16, 10))

# 將 2x3 的矩陣展開成一維，方便迴圈讀取
axes = axes.flatten()

for i, col in enumerate(cols_to_plot):
    sns.regplot(x='quality', y=col, data=df_white, ax=axes[i], 
                scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
    axes[i].set_title(f'{col} vs Quality')

plt.tight_layout()
plt.show()
```

```python
# 讀取兩個檔案
df_red = pd.read_csv('winequality-red.csv', sep=';')
df_white = pd.read_csv('winequality-white.csv', sep=';')

# 新增一欄標記類型
df_red['type'] = 'red'
df_white['type'] = 'white'

# 垂直合併
df_wine = pd.concat([df_red, df_white], axis=0, ignore_index=True)

print(f"紅酒資料筆數: {len(df_red)}")
print(f"白酒資料筆數: {len(df_white)}")
print(f"合併後總筆數: {len(df_wine)}")
```
